```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
```

```python
tickers=['^GSPC','BA','JPM','CCL','C','STX','WELL','EPR','XOM','HP','INTC','IBM','KSS',
         'LUV','WFC','CC','WYNN','FEYE','ENPH','OPEN','MSFT','FRO','SBUX','T','NAT','IVR','NL'
        ]

sec_prices=pd.DataFrame()
for t in tickers:
    print(t)
    sec_prices[t]=wb.DataReader(t,data_source='yahoo',start='2016-1-1',end='2020-12-31')['Adj Close']
sec_prices
```

```python
sec_returns=np.log(sec_prices/sec_prices.shift(1))
sec_returns
```

```python
sec_cov_with_market=sec_returns.cov()*250
sec_cov_with_market
```

```python
market_var=sec_returns['^GSPC'].var()*250
market_var
```

```python
beta_matrix=sec_cov_with_market/market_var
beta_matrix
betas=beta_matrix['^GSPC']
betas
```

```python
expected_returns=0.025+betas*0.05
expected_returns
```

```python
sharps=(expected_returns-0.025)/(sec_returns.std()*250**0.5)
sharps
```
