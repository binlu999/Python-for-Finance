```python
from pandas_datareader import data as wb
import numpy as np
import matplotlib.pyplot as plt
```

```python
PG=wb.DataReader("PG",data_source="yahoo",start="1995-1-1")
PG
```

## Log Returns

$$
ln(\frac{P_t}{P_(t-1)})
$$

```python
PG["log_return"]=np.log(PG["Adj Close"]/PG["Adj Close"].shift(1))
PG
```

```python
PG["log_return"].plot(figsize=(8,5))
plt.show()
```

```python
log_return_average_daily=PG["log_return"].mean()
log_return_average_daily
```

```python
log_return_average_annually=PG["log_return"].mean()*250
log_return_average_annually
```

```python
str(round(PG["log_return"].mean()*250,5)*100)+"%"
```
