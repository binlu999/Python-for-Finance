```python
import pandas as pd
from pandas_datareader import data as wb
import numpy as np
```

```python
tickers=['BA','JPM','CCL','C','STX','WELL','EPR','XOM','HP','INTC','IBM','KSS',
         'LUV','WFC','CC','WYNN','FEYE','ENPH','OPEN','MSFT','FRO','SBUX','T','NAT','IVR','NL'
        ]

weights=np.array([0.04,0.04,0.04,0.04,0.04,
        0.04,0.04,0.04,0.04,0.04,
        0.04,0.04,0.04,0.04,0.04,
        0.04,0.04,0.04,0.04,0.04,
        0.04,0.04,0.03,0.03,0.03,
        0.03])
```

```python
ds=pd.DataFrame()
for t in tickers:
    print(t)
    ds[t]=wb.DataReader(t,data_source='yahoo', start='2007-1-1')['Adj Close']

ds
```

```python
sec_returns_daily=np.log(ds/ds.shift(1))
sec_returns_daily
```

```python
sec_variance_annual=sec_returns_daily.var()*250
sec_variance_annual
```

```python
sec_covariance_matrix=sec_returns_daily.cov()*250
sec_covariance_matrix
```

```python
sec_corr_matrix=sec_returns_daily.corr()
sec_corr_matrix
```

```python
portfolio_variance=np.dot(weights.T,np.dot(sec_returns_daily.cov()*250,weights))
portfolio_variance
print(f'{round(portfolio_variance,5)*100}%')
```

```python
portfolio_volatility=portfolio_variance**0.5
print(f'{round(portfolio_volatility,5)*100}%')
```

```python
portfolio_diversifiable_risk=portfolio_variance - np.dot((weights**2).T, sec_variance_annual)
print(f'{round(portfolio_diversifiable_risk,3)*100} %')
```

```python
portfolio_non_diversifiable_risk=np.dot((weights**2).T,sec_variance_annual)
print(f'{round(portfolio_non_diversifiable_risk,5)*100} %')
```
