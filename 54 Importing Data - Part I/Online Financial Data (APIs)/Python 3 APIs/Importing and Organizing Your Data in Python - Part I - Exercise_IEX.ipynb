b. Extract data from IEX for Ford from the 1st of January 2015. The ticker you need is ‘F’.

```python
PG=wb.DataReader('PG',data_source='yahoo',start='2015-1-1')
```

```python
tickers=['PG','MSFT','T','F','GE']
new_data=pd.DataFrame()
for t in tickers:
    new_data[t]=wb.DataReader(t,data_source='yahoo', start='2015-1-1')['Close']
```
